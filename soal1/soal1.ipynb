```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data.csv', usecols=[2, 3, 7, 8])
df.head()
```

```python
target = df[df['Age']<=25]
target = target[target['Overall']>=80]
target = target[target['Potential'] >= 80]
target['status'] = 1
target.head()
```

```python
non_target = df.drop(target.index)
non_target['status'] = 0
non_target.head()
```

```python
train = pd.concat([target, non_target], axis=0).drop(columns="Name")
len(train)
```

```python
status = ['non_target', 'target']
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

def k():
    k = round(len(train) ** .5)
    if k % 2 == 0:
        return k + 1
    else:
        return k 
              
print(cross_val_score(
    KNeighborsClassifier(n_neighbors=k()), train[['Age', 'Overall', 'Potential']], train['status'],
    cv = 5
))
print(cross_val_score(
    RandomForestClassifier(n_estimators=100), train[['Age', 'Overall', 'Potential']], train['status'],
    cv = 5
))
print(cross_val_score(
    DecisionTreeClassifier(), train[['Age', 'Overall', 'Potential']], train['status'],
    cv = 5
))            
```
